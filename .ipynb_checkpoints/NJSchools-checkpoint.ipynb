```python
# Dependencies
import requests
import json
```

```python
# URLs 
url_elementary = "https://api.schooldigger.com/v1.2/rankings/schools/NJ?level=Elementary&perPage=50&appID=4199a348&appKey=5d16139d9ca9a3a48f2e4d3ac749468f"
url_high_schools = "https://api.schooldigger.com/v1.2/rankings/schools/NJ?level=High&appID=03b8de1a&appKey=%203172150ea0cd192a3e87274e65855c16"
```

```python
# Pretty Print the output of the JSON (Elementery School)
response_elementary = requests.get(url_elementary).json()
print(json.dumps(response_elementary, indent=4, sort_keys=True))
```

```python
# Pretty Print the output of the JSON (High School)
response_high_school = requests.get(url_high_schools).json()
print(json.dumps(response_high_school, indent=4, sort_keys=True))
```
