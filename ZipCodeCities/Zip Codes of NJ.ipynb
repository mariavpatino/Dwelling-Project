```python
zipCodes= pd.read_html("https://www.zipcodestogo.com/New%20Jersey/")
zipCodes[1]
```

```python
NJZip = zipCodes[1]
NJZip = NJZip[2:]
NJZip = NJZip.rename(columns={0:"Zip Code", 1:"City", 2:"County"})
NJZip.set_index("Zip Code",drop=True, inplace=True)
NJZip.drop(columns=3,inplace=True)
NJZip
```

```python
NJCities = NJZip['City'].unique().tolist()
NJCities
```

```python
cityLatLongURL = "https://public.opendatasoft.com/api/records/1.0/search/?dataset=us-zip-code-latitude-and-longitude&facet=state&facet=timezone&facet=dst&q="
# zipBoundaryURL = "https://nominatim.openstreetmap.org/search.php?polygon_geojson=1&format=json&q="
lats =[]
longs = []
count=0;
# bounds = []
for index,n in NJZip.iterrows():
#     print(f"{cityLatLongURL}{index}")
    temp = requests.get(f"{cityLatLongURL}{index}").json()
#     geojson_result = requests.get(f"{zipBoundaryURL}{index},US").json()
    count = count + 1
    try:
#         bounds.append(geojson_result[0]['geojson'])
#         lats.append(geojson_result[0]['lat'])
#         longs.append(geojson_result[0]['lon'])
        lats.append(temp["records"][0]['fields']['latitude'])
        longs.append(temp["records"][0]['fields']['longitude'])
    except:
#         bounds.append("NA")
        lats.append("NA")
        longs.append("NA")
        print(f"{count}: {index} couldn't be found")
#         lats.append("NA")
#         longs.append("NA")
#     print(temp)
    
```
